```python
# Ques 1

class NestedIterator:
    def __init__(self, nested_list):
        self.stack = []
        self.flatten(nested_list)

    def flatten(self, nested_list):
        for item in reversed(nested_list):
            if isinstance(item, list):
                self.flatten(item)
            else:
                self.stack.append(item)

    def hasNext(self):
        return len(self.stack) > 0

    def next(self):
        return self.stack.pop()
```

```python
# Ques 2

class Shape:
    def __init__(self, corners, density):
        self.corners = corners
        self.density = density
        self.area = self.compute_area()
        self.mass = self.compute_mass()
        self.mass_center = self.compute_mass_center()

    def compute_area(self):
        x_coords = [corner[0] for corner in self.corners]
        y_coords = [corner[1] for corner in self.corners]
        return abs(max(x_coords) - min(x_coords)) * abs(max(y_coords) - min(y_coords))

    def compute_mass(self):
        delta = 0.01
        mass = 0
        for x in range(int(min([corner[0] for corner in self.corners])), int(max([corner[0] for corner in self.corners])), int(delta)):
            for y in range(int(min([corner[1] for corner in self.corners])), int(max([corner[1] for corner in self.corners])), int(delta)):
                mass += delta ** 2 * self.density(x, y)
        return mass

    def compute_mass_center(self):
        delta = 0.01
        total_mass = self.mass
        xc = 0
        yc = 0
        for x in range(int(min([corner[0] for corner in self.corners])), int(max([corner[0] for corner in self.corners])), int(delta)):
            for y in range(int(min([corner[1] for corner in self.corners])), int(max([corner[1] for corner in self.corners])), int(delta)):
                xc += delta ** 2 * self.density(x, y) * x
                yc += delta ** 2 * self.density(x, y) * y
        xc /= total_mass
        yc /= total_mass
        return xc, yc
```

```python
# Ques 3

def tic_tac_toe(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != 'E':
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != 'E':
            return board[0][i]
    if board[0][0] == board[1][1] == board[2][2] != 'E':
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != 'E':
        return board[0][2]
    return "Draw"
```

```python
# Ques 4

import re

def remove_virus(files):
    cleaned_files = re.sub(r'\bvirus\.exe\b', '', files)
    cleaned_files = re.sub(r'\bdangerousvirus\.exe\b', '', cleaned_files)
    cleaned_files = re.sub(r'\blethalmalware\.exe\b', '', cleaned_files)
    return cleaned_files
```

```python
# Ques 5

def will_hit(trajectory, position):
    x, y = position
    equation = trajectory.split('=')[1].strip()
    m, b = equation.split('x')
    m = int(m.strip())
    b = int(b.strip())
    return y == m * x + b
```
