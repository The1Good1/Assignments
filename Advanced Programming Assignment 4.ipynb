```python
# Ques 1

def fastFib(num):
    if num <= 1:
        return num
    else:
        return fastFib(num - 1) + fastFib(num - 2)
```

```python
# Ques 2

def convert_to_hex(string):
    hex_values = []
    for char in string:
        hex_values.append(hex(ord(char))[2:])
    return ' '.join(hex_values)

```

```python
# Ques 3

def uncensor(censored_string, vowels):
    uncensored_string = ''
    vowel_index = 0
    for char in censored_string:
        if char == '*':
            uncensored_string += vowels[vowel_index]
            vowel_index += 1
        else:
            uncensored_string += char
    return uncensored_string

```

```python
# Ques 4

import socket

def get_domain(ip_address):
    try:
        domain = socket.gethostbyaddr(ip_address)[0]
        return domain
    except socket.herror:
        return "Unknown domain"
```

```python
# Ques 5

import math

def fact_of_fact(n):
    result = 1
    for i in range(1, n + 1):
        result *= math.factorial(i)
    return result
```
