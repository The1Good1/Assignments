```python
def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)
    
    # Initialize the 2D array dp with zeros
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill the first row with the ASCII sum of s2
    for j in range(1, n + 1):
        dp[0][j] = dp[0][j-1] + ord(s2[j-1])
    
    # Fill the first column with the ASCII sum of s1
    for i in range(1, m + 1):
        dp[i][0] = dp[i-1][0] + ord(s1[i-1])
    
    # Fill the remaining positions in dp
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                # Characters are equal, no deletion required
                dp[i][j] = dp[i-1][j-1]
            else:
                # Choose the minimum sum between deleting from s1 and deleting from s2
                dp[i][j] = min(dp[i-1][j] + ord(s1[i-1]), dp[i][j-1] + ord(s2[j-1]))
    
    # Return the minimum sum at the last position
    return dp[-1][-1]
```

```python
def checkValidString(s):
    stack = []
    wildcard_stack = []

    for c in s:
        if c == '(':
            stack.append(c)
        elif c == '*':
            wildcard_stack.append(c)
        elif c == ')':
            if stack:
                stack.pop()
            elif wildcard_stack:
                wildcard_stack.pop()
            else:
                return False

    # Match remaining opening parentheses with wildcards
    while stack and wildcard_stack:
        if stack[-1] < wildcard_stack[-1]:
            stack.pop()
            wildcard_stack.pop()
        else:
            break

    return len(stack) == 0
```

```python
def minDistance(word1, word2):
    m, n = len(word1), len(word2)
    
    # Create a 2D DP table to store the lengths of LCS
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    # Calculate the number of steps required
    lcs_length = dp[m][n]
    steps = m + n - 2 * lcs_length
    
    return steps

```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def constructBinaryTree(s):
    if not s:
        return None

    # Find the index of the first '('
    idx = s.find('(')

    if idx == -1:
        # No '(' found, s contains only the root value
        return TreeNode(int(s))

    # Create the root node with the value before '('
    root = TreeNode(int(s[:idx]))

    # Find the indices of the parentheses enclosing the left and right subtrees
    left_start = idx
    left_end = findMatchingParenthesis(s, left_start)
    right_start = left_end + 2  # Skip '(' and ')'
    right_end = findMatchingParenthesis(s, right_start)

    # Recursively construct the left and right subtrees
    root.left = constructBinaryTree(s[left_start + 1 : left_end])
    root.right = constructBinaryTree(s[right_start + 1 : right_end])

    return root


def findMatchingParenthesis(s, start):
    count = 0
    for i in range(start, len(s)):
        if s[i] == '(':
            count += 1
        elif s[i] == ')':
            count -= 1
            if count == 0:
                return i

    return -1


# Test the function
s = "4(2(3)(1))(6(5))"
root = constructBinaryTree(s)

# Function to print the binary tree in inorder traversal
def inorderTraversal(node):
    if node is None:
        return
    inorderTraversal(node.left)
    print(node.val, end=" ")
    inorderTraversal(node.right)


# Print the constructed binary tree
inorderTraversal(root)
```

```python
def compress(chars):
    # Initialize the pointers and count
    read = 0
    write = 0
    count = 1

    # Iterate through the characters
    for i in range(1, len(chars) + 1):
        if i < len(chars) and chars[i] == chars[i - 1]:
            # Found a repeating character, increment the count
            count += 1
        else:
            # Found the end of a group of repeating characters
            chars[write] = chars[read]
            write += 1
            if count > 1:
                # Append the count as characters
                count_str = str(count)
                for j in range(len(count_str)):
                    chars[write] = count_str[j]
                    write += 1
            # Reset the count
            count = 1
            read = i

    return write


# Test the function
chars = ["a", "a", "b", "b", "c", "c", "c"]
new_length = compress(chars)

# Print the compressed characters
print(chars[:new_length])
```

```python
from collections import defaultdict

def findAnagrams(s, p):
    result = []
    freq_p = defaultdict(int)
    freq_s = defaultdict(int)
    
    # Store the frequency of characters in p
    for char in p:
        freq_p[char] += 1
    
    # Initialize the sliding window
    window_start = 0
    window_end = 0
    
    # Iterate through s using the sliding window
    while window_end < len(s):
        # Expand the window by including the character at window_end
        freq_s[s[window_end]] += 1
        
        # Shrink the window if its size exceeds len(p)
        if window_end >= len(p):
            freq_s[s[window_start]] -= 1
            if freq_s[s[window_start]] == 0:
                del freq_s[s[window_start]]
            window_start += 1
        
        # Compare the frequency arrays
        if freq_p == freq_s:
            result.append(window_start)
        
        # Move the sliding window
        window_end += 1
    
    return result


```

```python
def decodeString(s):
    stack = []
    result = ""
    repetition_count = 0

    for char in s:
        if char.isdigit():
            repetition_count = repetition_count * 10 + int(char)
        elif char == '[':
            stack.append((result, repetition_count))
            result = ""
            repetition_count = 0
        elif char == ']':
            substring, count = stack.pop()
            result = substring + count * result
        else:
            result += char

    return result

```

```python
ddef buddyStrings(s, goal):
    if len(s) != len(goal):
        return False

    if s == goal:
        # Check if there are any duplicate characters in s
        seen = set()
        for char in s:
            if char in seen:
                return True
            seen.add(char)
        return False

    differences = []
    for i in range(len(s)):
        if s[i] != goal[i]:
            differences.append(i)

    if len(differences) != 2:
        return False

    i, j = differences
    if s[i] == goal[j] and s[j] == goal[i]:
        return True

    return False
```
