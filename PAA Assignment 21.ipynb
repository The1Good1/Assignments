Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you
predict them at all?

A probability distribution is a mathematical function that describes the likelihood of different outcomes in a random experiment or event. It provides a set of probabilities associated with each possible outcome. The distribution specifies the probabilities for all possible values that a random variable can take. While the values in a probability distribution are random, the distribution itself follows certain patterns or rules based on the underlying probability model.
Even though the values in a probability distribution are random, we can make predictions about their behavior based on the distribution's properties and mathematical principles. Probability theory allows us to understand and quantify the likelihood of different outcomes, calculate expected values, and analyze the overall behavior of random variables. By studying the probability distribution, we can make informed predictions about the likelihood of certain events occurring and analyze the characteristics of random processes.

Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is
one? Why are the latter considered “good enough”?

There is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated from a physical process or phenomenon that is inherently unpredictable, such as radioactive decay or atmospheric noise. They provide an unpredictable and unbiased sequence of numbers.


On the other hand, pseudo-random numbers are generated using deterministic algorithms. Although they are not truly random, they exhibit statistical properties similar to random numbers. Pseudo-random number generators (PRNGs) use an initial value called a seed and a mathematical algorithm to produce a sequence of numbers that appear random. If the seed and algorithm are known, the same sequence of numbers can be reproduced.

Pseudo-random numbers are considered "good enough" for many applications because they pass statistical tests for randomness and provide a sufficiently random-like sequence. They have a uniform distribution and exhibit properties such as independence and repeatability. In most cases, the pseudo-random numbers generated by PRNGs are indistinguishable from true random numbers for practical purposes. However, for applications requiring high-security cryptography or simulations that rely on true randomness, true random numbers are necessary.

Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

The two main factors that influence the behavior of a "normal" probability distribution are the mean and the standard deviation. The mean (μ) represents the central tendency or average value of the distribution. It determines the location of the peak or center of the distribution.


The standard deviation (σ) represents the spread or variability of the distribution. It quantifies the average distance between each data point and the mean. A larger standard deviation indicates a wider spread of values, while a smaller standard deviation indicates a narrower range of values.

Together, the mean and standard deviation fully characterize a normal distribution and determine its shape and properties. In a normal distribution, about 68% of the data falls within one standard deviation of the mean, approximately 95% falls within two standard deviations, and around 99.7% falls within three standard deviations.

Q4. Provide a real-life example of a normal distribution.

An example of a real-life normal distribution is the distribution of heights of adult humans. If you were to measure the heights of a large population of adults and plot them on a graph, you would typically observe a bell-shaped curve resembling a normal distribution. The majority of heights would cluster around the average or mean height, with fewer individuals at the extremes of very tall or very short heights. This distribution follows the characteristics of a normal distribution, with the mean representing the average height and the standard deviation representing the variability in heights.

Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term, the behavior of a probability distribution can be unpredictable due to the random nature of individual outcomes. However, as the number of trials or observations increases, the distribution tends to converge to its expected behavior or theoretical properties. This phenomenon is described by the law of large numbers and the central limit theorem.



As the number of trials grows, the observed frequencies of outcomes tend to approach the probabilities predicted by the distribution. The distribution becomes more stable, and the relative frequencies of outcomes become closer to the expected probabilities. In other words, the more trials conducted, the more closely the actual results resemble the predicted probabilities and characteristics of the distribution.

Q8. What is the relationship between exponentiation and logarithms?

Exponentiation and logarithms are mathematical operations that are inversely related to each other.
Exponentiation involves raising a base number to a given exponent. For example, in the expression 2^3, 2 is the base and 3 is the exponent. It signifies multiplying the base number (2) by itself three times: 2 * 2 * 2 = 8.

Logarithms, on the other hand, represent the inverse operation of exponentiation. A logarithm determines the exponent to which a given base must be raised to obtain a specific value. In the expression log(base 2)(8), the base is 2, and the logarithm signifies determining the exponent to which 2 must be raised to obtain the value 8. In this case, log(base 2)(8) = 3.

In summary, exponentiation calculates the result of raising a base to an exponent, while logarithms determine the exponent required to obtain a specific value when a base is raised to that exponent.
