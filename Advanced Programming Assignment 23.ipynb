```python
# Ques 1

class Sudoku:
    def __init__(self, board_string):
        self.board = []
        for i in range(9):
            row = [int(board_string[i * 9 + j]) for j in range(9)]
            self.board.append(row)

    def get_row(self, n):
        return self.board[n]

    def get_col(self, n):
        return [self.board[i][n] for i in range(9)]

    def get_sqr(self, n, m=None):
        if m is None:
            # Get square by index
            start_row = (n // 3) * 3
            start_col = (n % 3) * 3
            return [
                self.board[start_row + i][start_col + j]
                for i in range(3)
                for j in range(3)
            ]
        else:
            # Get square by cell position
            start_row = (n // 3) * 3
            start_col = (m // 3) * 3
            return [
                self.board[start_row + i][start_col + j]
                for i in range(3)
                for j in range(3)
            ]
```

```python
# Ques 2

def add_two_numbers(l1, l2):
    dummy = ListNode(0)
    current = dummy
    carry = 0

    while l1 or l2 or carry:
        value = carry

        if l1:
            value += l1.val
            l1 = l1.next

        if l2:
            value += l2.val
            l2 = l2.next

        carry = value // 10
        value %= 10

        current.next = ListNode(value)
        current = current.next

    return dummy.next
```

```python
# Ques 3

class CoffeeShop:
    def __init__(self, name, menu):
        self.name = name
        self.menu = menu
        self.orders = []

    def add_order(self, item):
        for menu_item in self.menu:
            if menu_item["item"] == item:
                self.orders.append(item)
                return "Order added!"
        return "This item is currently unavailable!"

    def fulfill_order(self):
        if self.orders:
            item = self.orders.pop(0)
            return f"The {item} is ready!"
        return "All orders have been fulfilled!"

    def list_orders(self):
        return self.orders

    def due_amount(self):
        total_amount = 0
        for order in self.orders:
            for menu_item in self.menu:
                if menu_item["item"] == order:
                    total_amount += menu_item["price"]
        return total_amount

    def cheapest_item(self):
        if self.menu:
            cheapest = min(self.menu, key=lambda x: x["price"])
            return cheapest["item"]
        return None

    def drinks_only(self):
        return [menu_item["item"] for menu_item in self.menu if menu_item["type"] == "drink"]

    def food_only(self):
        return [menu_item["item"] for menu_item in self.menu if menu_item["type"] == "food"]
```

```python
# Ques 4

def loneliest_number(lo, hi):
    def is_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

    def get_closest_prime(num):
        prime_before = None
        prime_after = None

        for i in range(num - 1, 1, -1):
            if is_prime(i):
                prime_before = i
                break

        for i in range(num + 1, hi + 1):
            if is_prime(i):
                prime_after = i
                break

        if prime_before is None:
            return prime_after
        elif prime_after is None:
            return prime_before
        else:
            distance_before = num - prime_before
            distance_after = prime_after - num
            if distance_before <= distance_after:
                return prime_before
            else:
                return prime_after

    lonely_number = None
    for num in range(lo, hi + 1):
        closest_prime = get_closest_prime(num)
        distance = abs(num - closest_prime)
        if lonely_number is None or distance > lonely_number['distance']:
            lonely_number = {
                'number': num,
                'distance': distance,
                'closest': closest_prime
            }

    return lonely_number


```

```python
# Ques 5

class Selfie:
    def __init__(self):
        self.state_history = []
        self.current_state = {}

    def save_state(self):
        self.state_history.append(self.current_state.copy())

    def recover_state(self, index):
        if index < 0 or index >= len(self.state_history):
            return self

        new_selfie = Selfie()
        new_selfie.state_history = self.state_history[:index + 1]
        new_selfie.current_state = self.state_history[index].copy()

        return new_selfie
```
