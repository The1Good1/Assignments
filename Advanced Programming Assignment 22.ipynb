```python
# Ques 1

class Smoothie:
    def __init__(self, ingredients):
        self.ingredients = ingredients

    def get_cost(self):
        ingredient_prices = {
            "Strawberries": 1.50,
            "Banana": 0.50,
            "Mango": 2.50,
            "Blueberries": 1.00,
            "Raspberries": 1.00,
            "Apple": 1.75,
            "Pineapple": 3.50
        }
        total_cost = sum(ingredient_prices.get(ingredient, 0) for ingredient in self.ingredients)
        return f"${total_cost:.2f}"

    def get_price(self):
        cost = self.get_cost()
        price = float(cost[1:]) + 1.5 * float(cost[1:])
        return f"${price:.2f}"

    def get_name(self):
        sorted_ingredients = sorted(self.ingredients)
        if len(sorted_ingredients) > 1:
            name = " ".join(sorted_ingredients) + " Fusion"
        else:
            name = sorted_ingredients[0] + " Smoothie"
        return name.replace("-berries", "-berry")
```

```python
# Ques 2

class Testpaper:
    def __init__(self, subject, markscheme, pass_mark):
        self.subject = subject
        self.markscheme = markscheme
        self.pass_mark = pass_mark


class Student:
    def __init__(self):
        self.tests_taken = "No tests taken"

    def take_test(self, testpaper, answers):
        score = sum(answer == markscheme for answer, markscheme in zip(answers, testpaper.markscheme))
        percentage = (score / len(testpaper.markscheme)) * 100
        result = "Passed" if percentage >= float(testpaper.pass_mark[:-1]) else "Failed"
        test_result = f"{result}! ({percentage:.0f}%)"
        if self.tests_taken == "No tests taken":
            self.tests_taken = {testpaper.subject: test_result}
        else:
            self.tests_taken[testpaper.subject] = test_result
```

```python
# Ques 3

class Train:
    def __init__(self, destinations, expected_time):
        self.destinations = destinations
        self.expected_time = expected_time


def manage_delays(train, destination, delay_minutes):
    if destination in train.destinations:
        hours, minutes = map(int, train.expected_time.split(':'))
        minutes += delay_minutes
        if minutes >= 60:
            hours += minutes // 60
            minutes %= 60
        train.expected_time = f"{hours:02d}:{minutes:02d}"
```

```python
class Minecart:
    def __init__(self, velocity):
        self.velocity = velocity


def mine_run(tracks):
    minecart = Minecart(0)
    for i, track in enumerate(tracks):
        if track == "-->":
            if minecart.velocity < 8:
                minecart.velocity += 2.67
        elif track == "<-->":
            pass
        elif track == "<--":
            if minecart.velocity > 0:
                minecart.velocity -= 2.67
                if minecart.velocity < 0:
                    minecart.velocity = 0
        elif track == "---":
            if minecart.velocity > 0:
                minecart.velocity -= 1
                if minecart.velocity < 0:
                    minecart.velocity = 0
        if minecart.velocity == 0:
            return i
    return True
```

```python
# Ques 5

class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height


def intersecting(rectangle1, rectangle2):
    if (
        rectangle1.x < rectangle2.x + rectangle2.width and
        rectangle1.x + rectangle1.width > rectangle2.x and
        rectangle1.y < rectangle2.y + rectangle2.height and
        rectangle1.y + rectangle1.height > rectangle2.y
    ):
        return True
    return False

```
